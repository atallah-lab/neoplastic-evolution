### Getting snap-velcro class probability distributions for each chromosome

```r
library(BSgenome.Hsapiens.UCSC.hg38)
k<-1
ENifrc <- 0.1
pds <- c()

for (chrnm in names(Hsapiens)[1:24]) {

    load(paste0("../data/root_maps/",chrnm,".rda"))
    
    map<-get(paste0(chrnm,"Map"))
    ict<-map[[2]]
    icl<-map[[3]]
    iot<-map[[4]]
    iol<-map[[5]]
    insites<-map[[1]]

    pd <- c(7.46*length(which(!is.na(ict))),
            3.48*length(which(!is.na(icl))),
            2.05*length(which(!is.na(iot))),
            1.00*length(which(!is.na(iol))))
    pd <- (pd/sum(pd))*(1-ENifrc)
    pd <- append(pd,ENifrc)
    
    pds[[k]]<-pd
    k<<-k+1
    
}

names(pds) <- names(Hsapiens)[1:24]
```
